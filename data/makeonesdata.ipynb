```python
import numpy as np

nsrcs = 5
nlens = 10
nbins = 20
ncombo = int(nsrcs*(nsrcs+1)/2 + nlens*nsrcs + nlens)
ndv = nbins*ncombo
print('ncombo = ', ncombo)
print('ndv = ', ndv)
```

```python
# produce ones mask
out = np.zeros((ndv,2))
out[:, 0] = np.arange(0, ndv)
out[:, 1] = 1
np.savetxt('./lsst_ones.mask', out ,fmt=["%d", "%.6e"])
```

```python
# produce ones cov
out = np.zeros((int(ndv*(ndv+1)/2),10))
cnt = 0
for i in range(ndv):
    for j in range(i,ndv):
        out[cnt, 0] = i
        out[cnt, 1] = j
        if i==j:
            out[cnt, 8] = 1
        cnt+=1
print('cnt = ', cnt,'; len(out) = ', len(out)) 
        
np.savetxt('./cov_lsst_eyes', out ,fmt=["%d", "%d", "%.6e", "%.6e", "%d", "%d", "%d", "%d", "%.6e", "%.6e"])
```
